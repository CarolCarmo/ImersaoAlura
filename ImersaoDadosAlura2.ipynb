```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

url_dados = 'https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true'

dados = pd.read_csv(url_dados, compression = 'zip')

dados
```

```python
mapa = {'droga': 'composto'}
dados.rename(columns=mapa, inplace=True) #inplace=True troca o nome da coluna pra sempre
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set() #salva as configurações padrões do seaborn
plt.figure(figsize=(8,6)) #altera o tamanho da figura (aumenta nesse caso), figure é uma função

#sns.countplot(x='composto', data=dados) #conta os dados e plota a frequencia total do dados
compostos = dados['composto'].value_counts().index[:5]  #index retorna o indice (nome do composto)

dados.query('composto in @compostos') #apenas os dados que forem iguais aos 5 indices de maior frequência definidos acima

ax = sns.countplot(x='composto', data=dados.query('composto in @compostos'), order=compostos) #countplot gera o gráfico, order ordena os compostos do mais frequente para o menos DESAFIO 1

ax.set_title('Top 5 compostos', fontsize=20) #set_title coloca o titulo do gráfico
ax.set_ylabel('Frequência do composto', fontsize=15) #DESAFIO 2, alterando legenda do eixo y
ax.set_xlabel('Nome do composto', fontsize=15) #legenda do eixo x

plt.show() #some os textos padrões da plotagem
```

```python
#dados[['g-0', 'g-1']] #slice das colunas que quero pegar
dados.loc[:,'g-0':'g-771'].describe() # [:,'g-0' ... pega todos os elementos de g-0 do primeiro ao ultimo elemento e a função toda pega todos os G's no describe
```

```python
dados.loc[:,'g-0':'g-771'].describe().T['mean'].hist(bins=30) #.T transpõe a matriz (transforma linhas em colunas), ploto no histograma a média
plt.title('Valores médios g')
plt.xlabel('Média')
plt.show()

dados.loc[:,'g-0':'g-771'].describe().T['min'].hist(bins=30) #.T transpõe a matriz (transforma linhas em colunas), ploto no histograma os valores minimos
plt.title('Valores mínimos g')
plt.xlabel('Mínimo')
plt.show()

dados.loc[:,'g-0':'g-771'].describe().T['max'].hist(bins=30) #.T transpõe a matriz (transforma linhas em colunas), ploto no histograma  os valores máximos
plt.title('Valores máximos g')
plt.xlabel('Máximo')
plt.show()
```
